```python
data['Average_marks']=data['math score']+data['reading score']+data['writing score']/3
data
```

# Question 2

```python
sns.countplot(x=data['parental level of education'])
plt.xticks(rotation=90)
plt.show
```

```python
plt.bar(data['parental level of education'],data['Average_marks'])
plt.xlabel('parental level of education')
plt.ylabel('Average_marks')
plt.xticks(rotation=90)
plt.show()
```

```python
#There is no impact on parental level of education and their children's average marks.
```
